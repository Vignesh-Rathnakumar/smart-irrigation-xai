```python
import pandas as pd
import numpy as np
import os

# -----------------------------------------
# STEP 1 — Load Climate Data
# -----------------------------------------

project_root = os.path.abspath("..")
data_folder = os.path.join(project_root, "data")

climate_path = os.path.join(data_folder, "multi_location_climate_data.csv")
soil_path = os.path.join(data_folder, "soil_data.csv")

climate_df = pd.read_csv(climate_path)
soil_df = pd.read_csv(soil_path)

print("Climate shape:", climate_df.shape)
print("Soil shape:", soil_df.shape)


# -----------------------------------------
# STEP 2 — Merge Climate + Soil
# -----------------------------------------

merged_df = pd.merge(
    climate_df,
    soil_df,
    on=["Latitude", "Longitude"],
    how="left"
)

print("Merged shape:", merged_df.shape)


# -----------------------------------------
# STEP 3 — Convert Date
# -----------------------------------------

merged_df["Date"] = pd.to_datetime(merged_df["Date"])


# -----------------------------------------
# STEP 4 — Compute ET (Evapotranspiration)
# -----------------------------------------

merged_df["ET0"] = (
    0.5 * merged_df["Temperature"]
    - 0.3 * (merged_df["Humidity"] / 100)
)

merged_df["Water_mm"] = merged_df["ET0"] - merged_df["Rainfall"]

merged_df["Water_mm"] = merged_df["Water_mm"].apply(lambda x: max(0, x))


# -----------------------------------------
# STEP 5 — Convert to Liters per Hectare
# -----------------------------------------

merged_df["Water_Liters"] = merged_df["Water_mm"] * 10000


# -----------------------------------------
# STEP 6 — Apply Soil Retention Factor
# -----------------------------------------

merged_df["Water_Liters"] = (
    merged_df["Water_Liters"] * merged_df["Soil_Factor"]
)


# -----------------------------------------
# STEP 7 — Add Season Feature
# -----------------------------------------

def get_season(month):
    if month in [1, 2]:
        return "Winter"
    elif month in [3, 4, 5]:
        return "Summer"
    elif month in [6, 7, 8, 9]:
        return "SW_Monsoon"
    else:
        return "NE_Monsoon"

merged_df["Season"] = merged_df["Date"].dt.month.apply(get_season)


# -----------------------------------------
# STEP 8 — Final Clean Dataset
# -----------------------------------------

final_df = merged_df[
    [
        "Date",
        "Latitude",
        "Longitude",
        "Temperature",
        "Humidity",
        "Rainfall",
        "Season",
        "Soil_Type",
        "Water_Liters"
    ]
]

print("Final dataset shape:", final_df.shape)


# -----------------------------------------
# STEP 9 — Save Final Dataset
# -----------------------------------------

final_path = os.path.join(data_folder, "final_spatial_irrigation_dataset.csv")
final_df.to_csv(final_path, index=False)

print("✅ Final dataset saved at:", final_path)

final_df.head()
```
