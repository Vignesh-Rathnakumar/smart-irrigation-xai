```python
import pandas as pd
import requests
import os
import time

# ----------------------------------------
# STEP 1 — Load Grid Points
# ----------------------------------------

project_root = os.path.abspath("..")
data_folder = os.path.join(project_root, "data")

grid_path = os.path.join(data_folder, "tamilnadu_grid_points.csv")
grid_df = pd.read_csv(grid_path)

print("Loaded grid points:", len(grid_df))


# ----------------------------------------
# STEP 2 — Define Time Range
# ----------------------------------------

start = "20200101"
end = "20241231"


# ----------------------------------------
# STEP 3 — Loop Through Each Location
# ----------------------------------------

all_data = []

for index, row in grid_df.iterrows():
    
    latitude = row["Latitude"]
    longitude = row["Longitude"]
    
    print(f"Fetching climate for: {latitude}, {longitude}")
    
    url = (
        "https://power.larc.nasa.gov/api/temporal/daily/point?"
        f"parameters=T2M,RH2M,PRECTOTCORR"
        f"&community=AG"
        f"&longitude={longitude}"
        f"&latitude={latitude}"
        f"&start={start}"
        f"&end={end}"
        f"&format=JSON"
    )
    
    try:
        response = requests.get(url, timeout=30)
        data = response.json()
        
        climate_data = data["properties"]["parameter"]
        
        df = pd.DataFrame({
            "Date": climate_data["T2M"].keys(),
            "Temperature": climate_data["T2M"].values(),
            "Humidity": climate_data["RH2M"].values(),
            "Rainfall": climate_data["PRECTOTCORR"].values()
        })
        
        df["Latitude"] = latitude
        df["Longitude"] = longitude
        
        all_data.append(df)
        
        time.sleep(1)  # prevent API overload
        
    except Exception as e:
        print("Error at location:", latitude, longitude)
        print(e)


# ----------------------------------------
# STEP 4 — Combine All Locations
# ----------------------------------------

master_df = pd.concat(all_data, ignore_index=True)

master_df["Date"] = pd.to_datetime(master_df["Date"])

print("Final dataset shape:", master_df.shape)


# ----------------------------------------
# STEP 5 — Save to Project Data Folder
# ----------------------------------------

output_path = os.path.join(data_folder, "multi_location_climate_data.csv")
master_df.to_csv(output_path, index=False)

print("✅ Climate dataset saved at:", output_path)

master_df.head()
```
